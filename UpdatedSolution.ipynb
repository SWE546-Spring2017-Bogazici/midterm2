## Error equations is

\begin{eqnarray}
\left(
\begin{array}{c}
\epsilon_1 \\ \vdots \\ \vdots \\ \epsilon_n 
\end{array}
\right)
=
\left(
\begin{array}{cc}
x_2 & -x_1 \\ x_3 & -x_2 \\ \vdots & \vdots \\ x_n & x_{N-1}  
\end{array}
\right) 
\left(
\begin{array}{c}
 1 \\ a  
\end{array}
\right)
\end{eqnarray}

## Minimizing errors
\begin{equation}
\frac{\partial \epsilon}{\partial a} = -2 \sum\limits_{i=2}^n x_i x_{i-1} + 2a \sum\limits_{i=1}^{n-1}x_i^2
\end{equation}

### a value is:

\begin{equation}
a = \frac{\sum\limits_{i=2}^n x_i x_{i-1}}{\sum\limits_{i=1}^{n-1} x_i^2}
\end{equation}

### Algorithm:

* Iterate over a and other unknowns for the problem until partial derivative is close to zero, figure out a value that sustains lowest possible error

```python
#initializing with randoms
p = random.random()
q = random.random()
r = random.random()
a = random.random()
limit = 100000
change = 0.0001

# Gradient solution a =  2.0327516806367942 , x_2 =  1.9804386720926568 , x_4 =  7.407777064202904 , x_6 =  30.062480740871607
#p = 1.9804386720926568
#q = 7.407777064202904
#r = 30.062480740871607
#a = 2.0327516806367942

y = [p,4,q,14.7,r,61.2] #given

def partialDerivative():
    return (-2*((y[1]*y[0])+(y[2]*y[1])+(y[3]*y[2])+(y[4]*y[3])+(y[5]*y[4])))+(2*a*(y[0]**2+y[1]**2+y[2]**2+y[3]**2+y[4]**2))

def calculateA():
    return ((y[1]*y[0])+(y[2]*y[1])+(y[3]*y[2])+(y[4]*y[3])+(y[5]*y[4]))/(y[0]**2+y[1]**2+y[2]**2+y[3]**2+y[4]**2)


def derivative_a(x,_x,a): #with respect to a
    return (2*a*_x**2)-(2*x*_x)
def derivate_x(x,_x,a): # with respect to x
    return (2*x)-(2*_x*a)
def derivate_x_(x,_x,a): # with respect to x_t-1
    return (2*a**2*_x)-(2*x*a)



A = [1,p,4,q,14.7,r]

def total_error():
    te = 0
    for i in range(len(y)):
        error = (A[i]*a) - y[i]
        error = round(error,6)
        te = te + error**2
    return te

a = calculateA()
for i in range(limit):
    p_plus = True
    partial = partialDerivative()
    if(p_plus):
        p = p+change
        y[0] = p
        tmpPartial = partialDerivative()
        if(tmpPartial > partial):
            p_plus = False
    else:
        p = p - change
        y[0] = p
        tmpPartial = partialDerivative()
        if(tmpPartial > partial):
            p_plus = True
    a = calculateA()
    q_plus = True
    partial = partialDerivative()
    if(q_plus):
        q = q+change
        y[2] = q
        tmpPartial = partialDerivative()
        if(tmpPartial > partial):
            q_plus = False
    else:
        q = q - change
        y[2] = q
        tmpPartial = partialDerivative()
        if(tmpPartial > partial):
            q_plus = True
    a = calculateA()
    r_plus = True
    partial = partialDerivative()
    if(r_plus):
        r = r+change
        y[4] = r
        tmpPartial = partialDerivative()
        if(tmpPartial > partial):
            r_plus = False
    else:
        r = r - change
        y[4] = r
        tmpPartial = partialDerivative()
        if(tmpPartial > partial):
            r_plus = True
    a = calculateA()
    
    

trials = 150
epoch = 5000
eta = 0.00003366

#In the loop for each value, change the value of a,p,q and r at each iteration. Value of a is changed multiple
#times per iteration.
for t in range(trials):
    for e in range(epoch):
        a_der = derivative_a(p,1,a)        
        a = a - (eta*a_der)
        x_der = derivate_x(p,1,a)
        p = p -(eta*x_der)
        a_der = derivative_a(4,p,a)        
        a = a - (eta*a_der)
        x__der = derivate_x_(4,p,a)
        p = p - (eta*x__der)
        a_der = derivative_a(q,4,a)        
        a = a - (eta*a_der)
        x_der = derivate_x(q,4,a)
        q = q - (eta*x_der)
        a_der = derivative_a(14.7,q,a)        
        a = a - (eta*a_der)
        x__der = derivate_x_(14.7,q,a)
        q = q - (eta*x__der)
        a_der = derivative_a(r,14.7,a)        
        a = a - (eta*a_der)
        x_der = derivate_x(r,14.7,a)
        r = r - (eta*x_der)
        a_der = derivative_a(61.2,r,a)        
        a = a - (eta*a_der)
        x__der = derivate_x_(61.2,r,a)
        r = r - (eta*x__der)
        y[0] = p
        A[1] = p
        y[2] = q
        A[3] = q
        y[4]=r
        A[5]=r
        
a = calculateA()
print(a)
print('p=',y[0]) # x_1
print('q=',y[2]) # x_3
print('r=',y[4]) # x_5
print('list =' , y)
```
